```python
vec2d = np.array([1,2])
s1 = 2
s2 = .5
s3 = -1
```

```python
plt.plot([0, vec2d[0]], [0, vec2d[1]], 'bs:' , label='v')
plt.plot([0, s1*vec2d[0]], [0, s1*vec2d[1]], 'ro-' , label='v*s1')
plt.plot([0, s2*vec2d[0]], [0, s2*vec2d[1]], 'kp-' , label='v*s2')
plt.plot([0, s3*vec2d[0]], [0, s3*vec2d[1]], 'g*-' , label='v*s3')

plt.axis('square')
plt.xlim([-4, 4])
plt.ylim([-4, 4])
plt.grid()
plt.legend()
plt.show()
```

```python
# Create 2 vectors(2D Vectors)
# Plot each vector
# plot (v1 + v2), (v1- v2), (v1*4 + v2/2)

v1 = np.random.randint(0, 10, size=(2))
v2 = np.random.randint(0, 10, size=(2))

plt.plot([0, v1[0]], [0, v1[1]], label='v1')
plt.plot([0, v2[0]], [0, v2[1]], label='v2')

plt.plot([0, v1[0] + v2[0]], [0, v1[1] + v2[1]], label='v1 + v2')
plt.plot([0, v1[0] - v2[0]], [0, v1[1] - v2[1]], label='v1 - v2')
plt.plot([0, (v1[0]*4 + v2[0]/2)], [0, (v1[1]*4 + v2[1]/2)], label='v1*4 + v2/2')

plt.legend()
plt.show()
```

```python
M1 = np.random.randint(0, 10, size=(5,2) )
M2 = np.random.randint(0, 10, size=(2,7))

print(M1)
print(' ')
print(M2)
print(' ')
print(np.matmul(M1, M2))
print(' ')
print(M1@M2)
print(' ')
print(np.matmul(M1, M2) - M1@M2)
```

```python
# Create a 3x3 of integers
# multiply by (1) identirty, (2) zeros, (3) that matrix's transpose

M = np.random.randint(1,10,size=(3,3))
print(M)
print(' ')

MI = np.eye(3)
print(MI)
print(' ')
print(M@MI)
print(' ')

print(' ')
print(M@np.zeros((3, 3)))
print(' ')

MT = np.transpose(M)
print(MT)
print(' ')
print(M@MT)
print(' ')
```

```python
A = np.random.randn(4,4)

Ainv = np.linalg.inv(A)

AinvA = A@Ainv

print(A)
print(' ')
print(Ainv)
print(A@Ainv)
```

```python
fig,ax = plt.subplots(1,3,figsize=(6,5))

ax[0].imshow(A)
ax[0].set_title('A')

ax[1].imshow(Ainv)
ax[1].set_title('A$^{-1}$') # latex

ax[2].imshow(AinvA)
ax[2].set_title('A$^{-1}$A') # latex

plt.show()
```
