```python
vec2d = np.array([1,2])
s1 = 2
s2 = .5
s3 = -1
```

```python
plt.plot([0, vec2d[0]], [0, vec2d[1]], 'bs:' , label='v')
plt.plot([0, s1*vec2d[0]], [0, s1*vec2d[1]], 'ro-' , label='v*s1')
plt.plot([0, s2*vec2d[0]], [0, s2*vec2d[1]], 'kp-' , label='v*s2')
plt.plot([0, s3*vec2d[0]], [0, s3*vec2d[1]], 'g*-' , label='v*s3')

plt.axis('square')
plt.xlim([-4, 4])
plt.ylim([-4, 4])
plt.grid()
plt.legend()
plt.show()
```

```python
# Create 2 vectors(2D Vectors)
# Plot each vector
# plot (v1 + v2), (v1- v2), (v1*4 + v2/2)

v1 = np.random.randint(0, 10, size=(2))
v2 = np.random.randint(0, 10, size=(2))

plt.plot([0, v1[0]], [0, v1[1]], label='v1')
plt.plot([0, v2[0]], [0, v2[1]], label='v2')

plt.plot([0, v1[0] + v2[0]], [0, v1[1] + v2[1]], label='v1 + v2')
plt.plot([0, v1[0] - v2[0]], [0, v1[1] - v2[1]], label='v1 - v2')
plt.plot([0, (v1[0]*4 + v2[0]/2)], [0, (v1[1]*4 + v2[1]/2)], label='v1*4 + v2/2')

plt.legend()
plt.show()
```

```python
# 1) Come up with 3 2D vectors, tow of them are orthogonal, but neither is orthogonal to the third
# 2) Plot all 3 vectors -- make sure the axes are square and equal

v1 = np.array([-4, 3])
v2 = np.array([3, 4])
v3 = np.array([6, 1])

plt.plot([0, v1[0]], [0, v1[1]], label='v1')
plt.plot([0, v2[0]], [0, v2[1]], label='v2')
plt.plot([0, v3[0]], [0, v3[1]], label='v3')

plt.axis('square')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.legend()
plt.show()
```
