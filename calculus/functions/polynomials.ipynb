```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
xDomain = [-5, 5]

coefficients = np.round(np.random.randn(4), 2)
x = np.linspace(xDomain[0], xDomain[1], 101)
y = np.zeros(len(x))

fname = '$y = '
for i, c in enumerate(coefficients):
    y += c*x**i
    fname += '+ '[int(c<0)] + f'{c:.2f}x^{i}'


plt.plot(x, y)
plt.xlim(xDomain[0], xDomain[1])
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.title(fname+ '$')
plt.grid()
plt.show()
```
