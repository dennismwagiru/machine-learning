```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

Implement the function and produce a graph
 $$\\y=x^2+3x^3-x^4$$


```python
xDomain = [-2, 2]

# option 1
x = np.linspace(xDomain[0], xDomain[1], 101)

# option 2
resolution = .1
x = np.arange(xDomain[0], xDomain[1]+ resolution, resolution)

y = x**2 + 3*x**3 - x**4

plt.plot(x, y, label='$y=x^2 + 3x^3 - x^4$')

plt.xlim(xDomain[0], xDomain[1])
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
s_beta = sym.var('beta')

# define the function
s_y = s_beta**2 + 3*s_beta**3 - s_beta**4

sym.plot(s_y, (s_beta, xDomain[0], xDomain[1]), xlabel='x', ylabel=None, title=f'$f(\\beta) = {sym.latex(s_y)}$')
```

```python
fx = sym.lambdify(s_beta, s_y)

yy = fx(x)

plt.plot(x, yy, label='$y=x^2 + 3x^3 - x^4$')
plt.grid()
plt.legend()
plt.xlim(xDomain[0], xDomain[1])
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.show()
```
