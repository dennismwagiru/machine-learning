```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplot(1,2,1)
plt.plot(np.random.randn(10))

plt.subplot(1,2,2)
plt.plot(np.random.randn(10))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8,5))

x = np.arange(10)
ax[0].plot(x, x**2, 'b')
ax[0].plot(x, x, 'k')
ax[1].plot(x, np.sqrt(x), 'r')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)

ax[0,0].plot(np.random.randn(4, 4))
ax[0,1].plot(np.random.randn(4, 4))
ax[1,0].plot(np.random.randn(4, 4))
ax[1,1].plot(np.random.randn(4, 4))

plt.tight_layout()
plt.show()
```

```python
# Exercise: Create a 3x3 subplot geometry in a for loop
fig, ax = plt.subplots(3,3)

for i in ax.flatten():
    for j in range(3):
        i.plot(np.random.randn(5,5))
        # print("{} - {}".format(i, j))
    

plt.tight_layout()
plt.show()
```
