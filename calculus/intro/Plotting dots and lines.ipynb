```python
import matplotlib.pyplot as plt

plt.plot(3,4, 'r^', label='red triangle')
plt.plot(2, 4.2, 'bo', label='blue dot')
plt.legend()
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 3*np.pi, 101)
y = np.sin(x)

plt.plot(x, y, 'm:')
plt.show()
```
