```python
import matplotlib.pyplot as plt
import numpy as np  
from imageio import imread
import seaborn as sns
```

```python
m = 3
n = 5

M = np.random.randint(10, size=(m, n))
print(M);

plt.imshow(M)

for i in range(m):
    for j in range(n):
        plt.text(j, i, str(M[i, j]), fontsize=20, horizontalalignment='center', 
        verticalalignment='center')
plt.colorbar()
plt.show()
```

```python
# Exercise: Visualise the Hilbert matrix using seaborn (heatmap)
# H is 10x10

n = 10
H = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        H[i, j] = 1 / ((i + 1) + (j+1) -1)

sns.heatmap(H)
plt.show()
```
